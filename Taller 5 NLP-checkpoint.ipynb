```python
import re
import numpy as np
import pandas as pd

from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')
from sklearn.feature_extraction.text import CountVectorizer

from nltk.stem.snowball import SnowballStemmer
spanishStemmer=SnowballStemmer("spanish")

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

# Punto 1: Pre-Procesamiento

- `[18 pts]` Leer el archivo `Princesas.csv` usando `pandas` y crear una nueva columna con el texto en minúscula, sin caracteres especiales ni números, sin palabras vacias y hacer stemming de las palabras 

```python
# Leer el archivo dialogos.csv usando pandas

import pandas as pd
import numpy as np

archivo = pd.read_csv("C:/Users/practicas/Especializacion/NLP/Talleres_NLP/Archivos Talleres/Princesas.csv")
archivo  
```

```python
def pre_procesado(texto):
    texto = texto.lower()
    texto = re.sub(r"[\W\d_]+", " ", texto)
    texto = " ".join([palabra for palabra in texto.split() if palabra not in stopwords_sp])  
    texto = " ".join([spanishStemmer.stem(palabra) for palabra in texto.split()])
    return texto

archivo['pre-procesado'] = archivo['Personalidad'].apply(lambda texto: pre_procesado(texto))
archivo
```

# Punto 2: TF-IDF

- `[16 pts]` Crear la matriz TF-IDF

```python
tfidf_vect = TfidfVectorizer()
tfidf = tfidf_vect.fit_transform(archivo['pre-procesado'].values)
tfidf_matrix = pd.DataFrame(tfidf.toarray(), columns=tfidf_vect.get_feature_names())
tfidf_matrix.index = archivo.Princesa
tfidf_matrix.T.round(3)

tfidf_vect = TfidfVectorizer()
tfidf = tfidf_vect.fit_transform(archivo['pre-procesado'].values)
tfidf_matrix = pd.DataFrame(tfidf.toarray(), columns=tfidf_vect.get_feature_names(), index= archivo.Princesa)
tfidf_matrix.T.round(3)
```
